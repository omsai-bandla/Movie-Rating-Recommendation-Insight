```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load ratings data

ratings = pd.read_csv(r"ratings.csv")   # Ensure ratings.csv is in the same folder

# 2. Convert timestamp  datetime
ratings["rating_dt"] = pd.to_datetime(ratings["timestamp"], unit="s")

# 3. Monthly average rating trend
monthly_trend = (
    ratings
    .groupby(pd.Grouper(key="rating_dt", freq="M"))["rating"]
    .mean()
    .reset_index()
)

# 4. Plot the trend
plt.figure(figsize=(12, 5))
plt.plot(monthly_trend["rating_dt"], monthly_trend["rating"], linewidth=2)

plt.xlabel("Month")
plt.ylabel("Average Rating")
plt.title("Average Movie Rating Trend Over Time")
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
ratings = pd.read_csv(r"ratings.csv")
movies  = pd.read_csv(r"movies.csv")

# Explode genres
movies_exp = movies.assign(genre=movies["genres"].str.split("|")).explode("genre")

# Merge ratings + genres
df = ratings.merge(movies_exp[["movieId", "title", "genre"]], on="movieId")

# Compute avg rating + count
stats = df.groupby(["genre", "title"]).rating.agg(["mean", "count"]).reset_index()

# Filter & take top 5
genre = "Action"    # <-- change genre here
top5 = (
    stats[(stats["genre"] == genre) & (stats["count"] >= 20)]
    .sort_values("mean", ascending=False)
    .head(5)
)

# Plot
plt.bar(top5["title"], top5["mean"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("Average Rating")
plt.title(f"Top 5 {genre} Movies")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
ratings = pd.read_csv(r"ratings.csv")
users  = pd.read_csv(r"users.csv")


# 1. Global average rating
global_avg = ratings["rating"].mean()

# 2. User-level stats
user_stats = (
    ratings.groupby("userId")["rating"]
    .agg(user_avg="mean", user_count="count", user_std="std")
    .reset_index()
)

# 3. Merge user info
user_stats = user_stats.merge(users, on="userId", how="left")

# 4. Difference from global average
user_stats["avg_diff_from_global"] = user_stats["user_avg"] - global_avg

# 5. Select users with enough ratings for analysis
filtered = user_stats[user_stats["user_count"] >= 20]

# ============================
# Plot 1: User Mean Difference vs Global Average
# ============================

plt.figure(figsize=(12, 5))
plt.scatter(filtered["userId"], filtered["avg_diff_from_global"])
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("User ID")
plt.ylabel("Difference From Global Average")
plt.title("Users' Rating Difference From Global Average")
plt.tight_layout()
plt.show()

# ============================
# Plot 2: User Rating Standard Deviation (Consistency)
# ============================

plt.figure(figsize=(12, 5))
plt.scatter(filtered["userId"], filtered["user_std"])
plt.xlabel("User ID")
plt.ylabel("Rating Standard Deviation")
plt.title("User Rating Consistency (STD)")
plt.tight_layout()
plt.show()
```
